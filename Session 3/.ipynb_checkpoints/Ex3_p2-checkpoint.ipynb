```python
fig, ax = plt.subplots()
x = np.linspace(-2,2,1000)
ax.plot(x, np.cosh(x), label ='cosh')
ax.plot(x, x**2/2, label = 'square/2')
ax.legend()
ax.grid()
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
plt.show()
```

```python
x = np.linspace(0.1,3,10)
y = x**3 + 2*x**2 -x -1
z = x**4
fig, ax = plt.subplots()
ax.scatter(x,y,s=z)
ax.set_xscale('log')
ax.grid()
plt.show()
```

```python
from scipy.interpolate import interp1d
A,nu,k = 10,4,2
def f(x,A,nu,k):
    return A*np.exp(-k*x)*np.cos(2*np.pi*nu*x)
x = np.linspace(0,0.5,8)
y= f(x,A,nu,k)
f_nearest = interp1d(x,y,kind='nearest')
f_linear = interp1d(x,y)
f_cubic = interp1d(x,y,kind='cubic')
```

```python
x2=np.linspace(0,0.5,100)
fig,ax=plt.subplots()
ax.scatter(x,y,label='datapoints')
ax.plot(x2,f_nearest(x2),label='nearest')
ax.plot(x2,f_linear(x2),label='linear')
ax.plot(x2,f_cubic(x2),label='cubic')
ax.legend()
plt.show()
```
