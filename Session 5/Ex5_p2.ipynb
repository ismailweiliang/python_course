```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime
```

```python
streamflow = pd.read_csv('Data/streamflow.txt',skiprows=30, delimiter='\t',
                        names=['Agency', 'SiteNumber','Date','Flow_cfs','Code']
                        ).iloc[1:]
streamflow.head()
```

```python
streamflow.dtypes
```

```python
streamflow['Flow_cfs'] = pd.to_numeric(streamflow['Flow_cfs'], errors='coerce')
streamflow.dtypes
```

```python
streamflow['Date'] = pd.to_datetime(streamflow['Date'])
streamflow.dtypes
```

```python
streamflow['Flow_cfs'].plot()
```

```python
streamflow = streamflow.set_index('Date')
streamflow.head()
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax.plot(streamflow.index, streamflow['Flow_cfs'],label='Flow')
ax.set_xlabel('Date')
ax.set_ylabel('Flow cfs')
ax.legend(loc='upper left')
plt.savefig('flowGraph.png')
plt.savefig('flowGraph.pdf')
plt.savefig('flowGraph.jpg')
plt.show()
```
