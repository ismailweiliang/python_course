```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime
import scipy.stats as st
```

```python
Yauyos = pd.read_csv('Data/Precip_Est_Yauyos.csv', index_col=0, parse_dates = True)
Yauyos.head()
```

```python
Yauyos.plot()
```

```python
promedio = Yauyos[Yauyos>0].mean()
desviacion = Yauyos[Yauyos>0].std()
promedio, desviacion
```

```python
tab = np.arange(Yauyos.Ppt.min(),Yauyos.Ppt.max(),0.1)
distnormal = st.norm.pdf(tab, loc=promedio, scale = desviacion)
distlognormal = st.pearson3.pdf(tab, skew=1, loc=promedio, scale=desviacion)
distweibull = st.dweibull.pdf(tab, c=1,loc=promedio, scale=desviacion)
distchi2 = st.chi2.pdf(tab, df=2, loc=promedio, scale=desviacion)

Yauyos[Yauyos>0].hist(bins=100,density=True)
plt.plot(tab, distnormal, label='Dist Normal')
plt.plot(tab, distlognormal, label='Dist Lognormal')
plt.plot(tab, distweibull, label='Dist Weibull')
plt.plot(tab, distchi2, label='Dist Chi2')
plt.legend()
```

```python
#T = 100y
#Probability of Highest event in 100 y = 1/100
#1 - 1/100 = 0.99
p100y_norm = st.norm.ppf([0.99], loc=promedio, scale=desviacion)
p100y_lognorm = st.pearson3.ppf([0.99], skew=1, loc=promedio, scale=desviacion)
p100y_weibull = st.dweibull.ppf([0.99], c=1, loc=promedio, scale=desviacion)
p100y_chi2 = st.chi2.ppf([0.99],df=2,loc=promedio, scale=desviacion)

print('Precoip for T=100y Normal Dist = %.2f'%p100y_norm[0])
print('Precoip for T=100y Log Normal Dist = %.2f'%p100y_lognorm[0])
print('Precoip for T=100y Weibull Dist = %.2f'%p100y_weibull[0])
print('Precoip for T=100y Chi2 Dist = %.2f'%p100y_chi2[0])
```

```python
p200y_norm = st.norm.ppf([0.995], loc=promedio, scale=desviacion)
p200y_lognorm = st.pearson3.ppf([0.995], skew=1, loc=promedio, scale=desviacion)
p200y_weibull = st.dweibull.ppf([0.995], c=1, loc=promedio, scale=desviacion)
p200y_chi2 = st.chi2.ppf([0.995],df=2,loc=promedio, scale=desviacion)

print('Precoip for T=200y Normal Dist = %.2f'%p200y_norm[0])
print('Precoip for T=200y Log Normal Dist = %.2f'%p200y_lognorm[0])
print('Precoip for T=200y Weibull Dist = %.2f'%p200y_weibull[0])
print('Precoip for T=200y Chi2 Dist = %.2f'%p200y_chi2[0])
```

```python
Yauyos['Year']= Yauyos.index.year
Yauyos.head()
```

```python
GroupedYauyos = Yauyos.groupby('Year').max()
GroupedYauyos.head()
```

```python
GroupedYauyos = GroupedYauyos.sort_values("Ppt",ascending=False)
GroupedYauyos.head()
```

```python
GroupedYauyos['Probability'] = [(i+1)/(len(GroupedYauyos)+1) for i
                               in range(len(GroupedYauyos))]
GroupedYauyos['Return Period'] = 1/GroupedYauyos['Probability']
GroupedYauyos.head()
```
